Simple Linear Regression memodelkan hubungan antara sebuah response variable dengan sebuah explanatory variable sebagai suatu garis lurus (linear)

### Sample Dataset

- Membuat pandas dataframe dengan menggunakan dictionary

```python
import pandas as pd

pizza = {'diameter': [6, 8, 10, 14, 18], 
         'harga': [7, 9, 13, 17.5, 18]}

pizza_df = pd.DataFrame(pizza)
pizza_df
```

### Visualisasi Data

- Gunakan scatter plot untuk membandingkan diameter dan harga, dimana jangkauan sumbu X dan Y aalah 25. 

```python
import matplotlib.pyplot as plt

pizza_df.plot(kind='scatter', x='diameter', y='harga')

plt.title('Perbandingan Diameter dan Harga Pizza')
plt.xlabel('Diameter (inch)')
plt.ylabel('Harga (dollar)')
plt.xlim(0, 25)
plt.ylim(0, 25)
plt.grid(True)
plt.show()
```

### Simple Linear Regression Model

#### Penyesuaian Dataset

- Kita dapat mengelompokan features dan target menggunakan fungsi `np.array()`
- Numpy array diameter masih berbentuk 1 dimensi, agar dapat digunakan dalam scikit-learn kita konversi menjadi 2 dimensi menggunakan method `reshape()`.

```python
import numpy as np

X = np.array(pizza_df['diameter'])
y = np.array(pizza_df['harga'])

print(f'X: {X}')
print(f'y: {y}')
```

```python
X = X.reshape(-1, 1)
X.shape
```

```python
X
```

#### Training Simple Linear Regression Model

- Untuk membuat object Simple Linear Regression Model, gunakan fungsi `LinearRegression()`.
- Lakukan training dengan method `fit()`.
    - Parameter 1 diisi dengan nilai features.
    - Parameter 2 diisi dengan nilai target.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

#### Visualisasi Simple Linear Regression Model

- Membuat ploting dengan panjang 25, kemudian di kenakan fungsi `reshape()` untuk menampung nilai features.
- Gunakan fungsi `predict()` untuk menampung nilai target berdasarkan nilai features.
- Buatlah plot menggunakan fungsi `plt.scatter()` dan tampilkan garis menggunakan fungsi `plt.plot()`.
- Fungsi `plt.xlim`/ `.ylim` digunakan untuk membatasi plot.
- Untuk menampilkan nilai intercept dapat menggunakan `.intercept_` dan `.coef_` untuk slope.

```python
X_vis = np.array([0, 25]).reshape(-1, 1)
y_vis = model.predict(X_vis)
```

```python
plt.scatter(X, y)
plt.plot(X_vis, y_vis, '-r')

plt.title('Perbandingan Diameter dan Harga Pizza')
plt.xlabel('Diameter (inch)')
plt.ylabel('Harga (dollar)')
plt.xlim(0, 25)
plt.ylim(0, 25)
plt.grid(True)
plt.show()
```

Formula Linear Regression: $y = \alpha + \beta x$ 
- $y$: response variable
- $x$: explanatory variable
- $\alpha$: intercept
- $\beta$: slope

```python
print(f'intercept: {model.intercept_}')
print(f'slope: {model.coef_}')
```

#### Mencari nilai slope

- Untuk mengkonverasi array 2 dimensi menjadi 1 dimensi gunakan fungsi `.flatten()`.

Nilai slope pada Linear Regression bisa diperoleh dengan memanfaatkan formula berikut:

$\beta = \frac{cov(x, y)}{var(x)}$

```python
print(f'X:\n{X}\n')
print(f'X flatten: {X.flatten()}\n')
print(f'y: {y}')
```

##### Variance

- Untuk menghitung nilai variance dapt menggunakan fungsi `np.var()`.
    - Parameter 2 berisi degree of freedem

```python
variance_x = np.var(X.flatten(), ddof=1)

print(f'variance: {variance_x}')
```

##### Covariance

- Untuk menghitung nilai covariance dapt menggunakan fungsi `np.cov()`.
    - Parameter 2 diisi dengan nilai y
- Untuk menentukan covarience yang akan kita gunakan dapat menggunakan notasi index/ [...]

```python
np.cov(X.flatten(), y)
```

```python
covariance_xy = np.cov(X.flatten(), y)[0][1]

print(f'covariance: {covariance_xy}')
```

##### Slope

- Untuk menghitung slope bagi covariance xy dengan variance x

```python
slope = covariance_xy / variance_x

print(f'slope: {slope}')
```

#### Mencari nilai intercept

Nilai intercept pada Linear Regression bisa diperoleh dengan memanfaatkan formula berikut:

$\alpha = \bar{y} - \beta \bar{x}$

- Nilai intercept merupakan hasil dari nilai rata-rata target dikurangi slope dikasli rata-rata features

```python
intercept = np.mean(y) - slope * np.mean(X)

print(f'intercept: {intercept}')
```

#### Prediksi Harga Pizza

- Membuat prediksi harga pizza dengan diameter 12, 20, 23
- Melakukan prediksi dengan memanggil fungsi `model.predict()` 

```python
diameter_pizza = np.array([12, 20, 23]).reshape(-1, 1)
diameter_pizza
```

```python
prediksi_harga = model.predict(diameter_pizza)
prediksi_harga
```

```python
for dmtr, hrg in zip(diameter_pizza, prediksi_harga):
    print(f'Diameter: {dmtr} predilsi harga: {hrg}')
```

#### Evaluasi Simple Linear Regression Model

##### Training & Testing Dataset

- Membuat training dan testing dataset untuk setiap target dan features

```python
X_train = np.array([6, 8, 10, 14, 18]).reshape(-1, 1)
y_train = np.array([7, 9, 13, 17.5, 18])

X_test = np.array([8, 9, 11, 16, 12]).reshape(-1, 1)
y_test = np.array([11, 8.5, 15, 18, 11])
```

##### Training Simple Linear Regression Model

- Untuk membuat object Simple Linear Regression Model, gunakan fungsi `LinearRegression()`.
- Lakukan training dengan method `fit()`.
    - Parameter 1 diisi dengan nilai features.
    - Parameter 2 diisi dengan nilai target.

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

##### Evaluasi Linear Regression Model dengan Coefficient of Determination atau R-squared ($R^2$)
Referensi: https://en.wikipedia.org/wiki/Coefficient_of_determination

- R-squared dapat digunakan untuk mengevaluasi LinearRegressionModel object
- Prediksi nilai features untuk mendapatkan nilai target
- Bandingkan nilai target dan nilai prediksi menggunakan fungsi `score()`.

```python
from sklearn.metrics import r2_score

y_pred = model.predict(X_test)

r_squared = r2_score(y_test, y_pred)

print(f'R-squared: {r_squared}')
```

#### Mencari nilai R-squared ($R^2$)

$R^{2} = 1 - \frac{SS_{res}}{SS_{tot}}$


$SS_{res} =  \sum_{i=1}^{n}(y_i - f(x_i))^2$


$SS_{tot} =  \sum_{i=1}^{n}(y_i - \bar{y})^2$

##### $SS_{res}$

```python
ss_res = sum([(y_i - model.predict(x_i.reshape(-1, 1))[0])**2
              for x_i, y_i in zip(X_test, y_test)])

print(f'ss_res: {ss_res}')
```

##### $SS_{tot}$

```python
mean_y = np.mean(y_test)
ss_tot = sum([(y_i - mean_y)**2 for y_i in y_test])

print(f'ss_tot: {ss_tot}')
```

##### $R^2$

```python
r_squared = 1 - (ss_res / ss_tot)

print(f'R-squared: {r_squared}')
```
